```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data= pd.read_csv('heart - Sheet5.csv')
```

```python
data.head()
```

```python
data
```

```python
data = data.replace('?', np.NaN)
data.head(10)
```

```python
data.isnull().values.any()
```

```python
missing = data.isnull().sum(axis=0) 
print("Columns and Number of missing data \n", missing)
```

```python
data['trestbps'] = data.trestbps.astype(float)
data['trestbps'].fillna((data['trestbps'].mean()), inplace=True)
data['trestbps'] = data.trestbps.astype(int)

data['chol'] = data.chol.astype(float)
data['chol'].fillna((data['chol'].mean()), inplace=True)
data['chol'] = data.chol.astype(int)

data['fbs'] = data.fbs.astype(float)
data['fbs'].fillna((data['fbs'].mean()), inplace=True)
data['fbs'] = data.fbs.astype(int)

data['restecg'] = data.restecg.astype(float)
data['restecg'].fillna((data['restecg'].mean()), inplace=True)
data['restecg'] = data.restecg.astype(int)

data['thalach'] = data.thalach.astype(float)
data['thalach'].fillna((data['thalach'].mean()), inplace=True)
data['thalach'] = data.thalach.astype(int)

data['exang'] = data.exang.astype(float)
data['exang'].fillna((data['exang'].mean()), inplace=True)
data['exang'] = data.exang.astype(int)

data['slope'] = data.slope.astype(float)
data['slope'].fillna((data['slope'].mean()), inplace=True)
data['slope'] = data.slope.astype(int)

data['ca'] = data.ca.astype(float)
data['ca'].fillna((data['ca'].mean()), inplace=True)
data['ca'] = data.ca.astype(int)

data['thal'] = data.thal.astype(float)
data['thal'].fillna((data['thal'].mean()), inplace=True)
data['thal'] = data.thal.astype(int)
```

```python
def plot_corr(data, size=11):
    
    
    corr = data.corr()    # data frame correlation function
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)   # color code the rectangles by correlation value
    plt.xticks(range(len(corr.columns)), corr.columns)  # draw x tick marks
    plt.yticks(range(len(corr.columns)), corr.columns)  # draw y tick marks
```

```python
plot_corr(data)
```

```python
num_obs = len(data)
num_true = len(data.loc[data['target'] == 1])
num_false = len(data.loc[data['target'] == 0])
print("Number of True cases:  {0} ({1:2.2f}%)".format(num_true, (num_true/num_obs) * 100))
print("Number of False cases: {0} ({1:2.2f}%)".format(num_false, (num_false/num_obs) * 100))
```

```python
X=data[['age','sex', 'cp' ,'trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','thal']]
Y=data['target']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,random_state=0)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
print('Test Accuracy:',classifier.score(X_test,y_test))
print('Train Accuracy:',classifier.score(X_train,y_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier()
classifier.fit(X_train,y_train)
print('Test Accuracy:',classifier.score(X_test,y_test))
print('Train Accuracy:',classifier.score(X_train,y_train))
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
print('Test Accuracy:',classifier.score(X_test,y_test))
print('Train Accuracy:',classifier.score(X_train,y_train))
```

```python
from sklearn import tree
classifier = tree.DecisionTreeClassifier()
classifier.fit(X_train,y_train)
print('Test Accuracy:',classifier.score(X_test,y_test))
print('Train Accuracy:',classifier.score(X_train,y_train))
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train,y_train)
print ("Training Accuracy {}".format(classifier.score(X_train,y_train)))
print ("Testing Accuracy {}".format(classifier .score(X_test,y_test)))
```

```python
def group_age(age):
    if age <= 30:
        return "<= 30"
    elif age <= 40:
        return "31-40"
    elif age<= 50:
        return "41-50"
    elif age<= 60:
        return "51-60"
    else:
        return "61-70"
```

```python
data["age"]= data.age.apply(group_age)
```
